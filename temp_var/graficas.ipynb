```python
import numpy as np
import matplotlib.pyplot as plt

def dih_energy(phi_deg):
    # Coeficientes originales (en kcal/mol)
    c0 = 1.736
    c1 = -4.490
    c2 = 0.776
    c3 = 6.990
    
    # Cálculo en kcal/mol (usando grados para el coseno)
    phi_rad = np.radians(phi_deg)
    cos_phi = np.cos(phi_rad)
    
    E = c0 + c1*cos_phi + c2*(cos_phi**2) + c3*(cos_phi**3)
    
    # Conversión a kJ/mol
    return E * 4.184

# Generar datos de 0 a 360 grados
phi_values = np.linspace(0, 360, 500)
energy_values = [dih_energy(p) for p in phi_values]

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(phi_values, energy_values, label='Potencial OPLS (PE)', color='blue', lw=2)
plt.axhline(0, color='black', linestyle='--', alpha=0.3)
plt.title('Perfil de Energía del Ángulo Diedro')
plt.xlabel('Ángulo Diedro (grados)')
plt.ylabel('Energía (kJ/mol)')
plt.grid(True, alpha=0.3)
plt.xticks(np.arange(0, 361, 60))
plt.legend()

plt.show()
```
